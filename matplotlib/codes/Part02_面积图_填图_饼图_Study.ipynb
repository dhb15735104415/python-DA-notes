```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 面积图 plt.plot.area()
# 面积图又称区域图，强调数量随时间而变化的程度，也可用于引起人们对总值趋势的注意
fig,axes = plt.subplots(2,1,figsize = (8,6))
df1 = pd.DataFrame(np.random.rand(10, 4), columns=list('abcd'))
df2 = pd.DataFrame(np.random.randn(10, 4), columns=list('abcd'))

df1.plot.area(ax=axes[0],colormap='Set1',alpha=0.6)
df2.plot.area(ax=axes[1],colormap='Set2',alpha=0.8,stacked=False)  # stacked必须为False

# 使用Series.plot.area()和DataFrame.plot.area()创建面积图
# stacked：是否堆叠，默认情况下，区域图被堆叠
# 为了产生堆积面积图，每列必须是正值或全部负值！
# 当数据有NaN时候，自动填充0，所以图标签需要清洗掉缺失值

plt.show()
```

```python
# 填图
fig,axes = plt.subplots(2,1,figsize=(10,6))

x = np.linspace(0,1,500)
y1 = np.sin(4*np.pi*x)*np.exp(-5*x)
y2 = -np.sin(4*np.pi*x)*np.exp(-5*x)
# np.exp()计算输入数组中所有元素的指数。

axes[0].fill(x,y1,'r',alpha=0.6,label='y1')
axes[0].fill(x,y2,'g',alpha=0.6,label='y2')

# 对函数与坐标轴之间的区域进行填充，使用fill函数
# 也可写成：
# plt.fill(x,y1,'r',x,y2,'g',alpha=0.6,label='test_plt_fill')

# 填充两个函数之间的区域，使用fill_between函数
x = np.linspace(0,5 * np.pi,100)
y1 = np.sin(x)
y2 = np.cos(x)
axes[1].fill_between(x,y1,y2,color='g',alpha=0.6,label='fill_between')

# 添加图列 网格
for i in range(2):
    axes[i].legend()
    axes[i].grid()
plt.show()
```

```python
# 饼图 plt.pie()
# plt.pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=None, 
# radius=None, counterclock=True, wedgeprops=None, textprops=None, center=(0, 0), frame=False, hold=None, data=None)

s = pd.Series(np.random.rand(4), index=list('abcd'), name='series')
plt.axis('equal')  # 保证长宽相等
# 绘制饼图
plt.pie(s,
        explode=[0.1,0,0,0],
        labels=s.index,
        colors=['r','g','lightblue','y'],
        autopct='%.2f%%',
        pctdistance=0.5,
        labeldistance=1.1,
        shadow=True,
        startangle=5,
        radius=1.2,
        frame=False
        )
# 第一个参数：数据
# explode：指定每部分的偏移量,可以理解为每块图之间的缝隙距离 
# labels：标签
# colors：颜色
# autopct：饼图上的数据标签显示方式
# pctdistance：每个饼切片的中心和通过autopct生成的文本开始之间的比例,默认:0.6
# labeldistance：绘制饼图标签的径向距离(直径),用来表示label(abcd)位置表示,默认值：1.1
# shadow：阴影
# startangle：开始角度,如果不是None，则将饼图的起点从x轴逆时针旋转* angle *度。
# radius：半径 如果* radius *为* None *，它将设置为1。
# frame：图框,如果为true，则用图表绘制轴框架。
# counterclock：指定指针方向，顺时针或者逆时针,默认为True(逆时针)
plt.show()
```

```python
# 设置子图布局 plt.subplot2grid()
# 生成3*3的子图
ax1 = plt.subplot2grid((3,3),(0,0))
ax2 = plt.subplot2grid((3,3),(1,0))  # 第2行第1个图
ax3 = plt.subplot2grid((3,3),(0,2),rowspan=3)  # 第1行第3个图 占据3格行,即高度为3
ax4 = plt.subplot2grid((3,3),(2,0),colspan=2)  # 第3行第1个图 占据2格列,即宽度为2
ax5 = plt.subplot2grid((3,3),(0,1),rowspan=2)  # 第1行第2个图 占据2格列,即宽度为2

plt.show()
```

```python
# 嵌套图,在图里嵌套子图
# 本质上就是在图里新建一个坐标系,并绘图

x = np.linspace(0,10,1000)
y1 = x**2
y2 = np.sin(x**2)

fig,ax1 = plt.subplots()  # 无参则构建创建一个图和一组子图(个数也为1)。
ax1.plot(x,y1)

# 在ax1图上嵌套一个子图
left,bottom,width,height = [0.22,0.45,0.3,0.4]
ax2 = fig.add_axes([left,bottom,width,height])  #  在图中添加轴 
ax2.plot(x,y2)

plt.show()
```
