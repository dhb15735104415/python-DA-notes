```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# 刻度

from matplotlib.ticker import MultipleLocator, FormatStrFormatter
t = np.arange(0.0, 100.0, 1)
s = np.sin(0.1*np.pi*t)*np.exp(-t*0.01)
ax = plt.subplot(111)  # subplot()在当前图形上添加一个子图
#注意:一般都在ax中设置,不再plot中设置

# 绘图,并设置线条 点的样式及颜色
plt.plot(t,s,'--g*')  
# 加网格线,x/y轴都加
plt.grid(True, linestyle = "--",color = "gray", linewidth = "0.5",axis = 'both')

#plt.legend(loc = 'upper left')  # 图例


xmajorLocator = MultipleLocator(10) # 将x主刻度标签设置为10的倍数
xmajorFormatter = FormatStrFormatter('%.0f') # 设置x轴标签文本的格式,%.0f表示整数
xminorLocator   = MultipleLocator(5) # 将x轴次刻度标签设置为5的倍数  
ymajorLocator = MultipleLocator(0.5) # 将y轴主刻度标签设置为0.5的倍数,小数点保留1位
ymajorFormatter = FormatStrFormatter('%.1f') # 设置y轴标签文本的格式
yminorLocator   = MultipleLocator(0.1) # 将此y轴次刻度标签设置为0.1的倍数  

ax.xaxis.set_major_locator(xmajorLocator)  # 设置x轴主刻度
ax.xaxis.set_major_formatter(xmajorFormatter)  # 设置x轴标签文本格式
ax.xaxis.set_minor_locator(xminorLocator)  # 设置x轴次刻度
 
ax.yaxis.set_major_locator(ymajorLocator)  # 设置y轴主刻度
ax.yaxis.set_major_formatter(ymajorFormatter)  # 设置y轴标签文本格式
ax.yaxis.set_minor_locator(yminorLocator)  # 设置y轴次刻度

ax.xaxis.grid(b =True, which='major') #x坐标轴的网格使用主刻度
ax.yaxis.grid(b =None, which='minor') #y坐标轴的网格使用次刻度
# y轴b=None, 但是有关键字参数which,因此网格依旧会显示

'''     
   b : bool or None
        Whether to show the grid lines. If any *kwargs* are supplied,
        it is assumed you want the grid on and *b* will be set to True.

        If *b* is *None* and there are no *kwargs*, this toggles the
        visibility of the lines.

    which参数：配置网格线,源码如下:
    which : {'major', 'minor', 'both'}
            major:主刻度
            minor:次刻度
            both:二者都配置
'''

#删除坐标轴的刻度显示
#ax.yaxis.set_major_locator(plt.NullLocator()) 
#ax.xaxis.set_major_formatter(plt.NullFormatter()) 

plt.show()
```
