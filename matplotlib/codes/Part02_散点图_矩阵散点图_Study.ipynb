```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# plt.scatter()散点图
# plt.scatter(x, y, s=20, c=None, marker='o', cmap=None, norm=None, vmin=None, vmax=None, 
#            alpha=None, linewidths=None, verts=None, edgecolors=None, hold=None, data=None, **kwargs)

x = np.random.rand(1000)
y = np.random.rand(1000)
plt.figure(figsize=(8,4))
plt.scatter(x, y,
            marker='^',
            s = x*100,
            c = x,
            cmap='Reds',
            alpha=0.6)
# s：散点的大小
# c：散点的颜色
# cmap：colormap,.Colormap实例或注册的颜色图名称。
#       仅当c参数是浮点数数组时，才使用* cmap *。如果为None，则默认为rc image.cmap。
# vmin,vmax：亮度设置，标量

plt.show()
```

```python
# pd.plotting.scatter_matrix散点矩阵
# pd.plotting.scatter_matrix(frame, alpha=0.5, figsize=None, ax=None, 
# grid=False, diagonal='hist', marker='.', density_kwds=None, hist_kwds=None, range_padding=0.05, **kwds)

df = pd.DataFrame(np.random.randn(100,4),columns=list('abcd'))
pd.plotting.scatter_matrix(df,
                           alpha=0.8,
                           figsize=(10,6),
                           marker='o',
                           diagonal='kde')

# diagonal:({‘hist’, ‘kde’})，必须且只能在{‘hist’, ‘kde’}中选择1个 → 每个指标的频率图,对角线位置图的显示是直方图还是核密度估计
# range_padding:(float, 可选)，图像在x轴、y轴原点附近的留白(padding)，该值越大，留白距离越大，图像远离坐标原点

plt.show()
```

```python
# 坐标值显示,即增加注释
x_coords = [0.13, 0.22, 0.39, 0.59, 0.68, 0.74, 0.93]
y_coords = [0.75, 0.34, 0.44, 0.52, 0.80, 0.25, 0.55]

plt.figure(figsize=(8,6))
plt.scatter(x_coords,y_coords,alpha=0.8,marker='o')

for x,y in zip(x_coords,y_coords):
    plt.annotate('(%s,%s)' %(x,y),xy=(x,y),xytext=(0,-15),textcoords='offset points',ha='left')

# ha:与坐标对齐的方式 'center left right'
plt.show()
```
