```python
''' 
    Numpy随机数

    numpy.random包含多种概率分布的随机样本，是数据分析辅助的重点工具之一
'''
```

```python
import numpy as np

# 随机数生成
print(np.random.normal(size=4*4))  # 生成一个标准正太分布的4*4样本值
```

```python
# numpy.random.rand(d0, d1, ..., dn)：生成一个[0,1)之间的随机浮点数或N维浮点数组 —— 均匀分布

a = np.random.rand()
print(a,type(a))  # 生成一个随机浮点数[0,1)

b = np.random.rand(4)  # 生成形状为4的一维数组
print('生成形状为4的一维数组',b)

c = np.random.rand(4,4)
print('生成形状为4*4的二维数组', c)   # 生成形状为4*4的二维数组

# 可以进行数学计算
print('乘以100后:',c*100)
```

```python
# numpy.random.randint(low, high=None, size=None, dtype='l')：生成一个整数或N维整数数组
# 若high不为None时，取[low,high)之间随机整数，否则取值[0,low)之间随机整数，且high必须大于low 
# dtype参数：只能是int类型 

# low=2：生成1个[0,2)之间随机整数
int1 = np.random.randint(2)
print('low=2：生成1个[0,2)之间随机整数:',int1)

# low=2,high=5 ：生成1个[2,5)之间随机整数
int2 = np.random.randint(2,5)
print('low=2,high=5 ：生成1个[2,5)之间随机整数:',int2)

# low=2,size=5 ：生成5个[0,2)之间随机整数
int_arr1 = np.random.randint(2,size=5)
print(f'low=2,size=5 ：生成5个[0,2)之间随机整数:{int_arr1},类型为:',type(int_arr1))

# low=2,high=6,size=8：生成8个[2,6)之间随机整数
int_arr2 = np.random.randint(2,6,size=8)
print('low=2,high=6,size=8：生成8个[2,6)之间随机整数:',int_arr2)

# low=2,size=(2,3)：生成一个2x3整数数组,取数范围：[0,2)随机整数 
int_arr3 = np.random.randint(2,size=(2,3))
print('low=2,size=(2,3)：生成一个2x3整数数组:',int_arr3)

# low=2,high=6,size=(2,3)：生成一个2*3整数数组,取值范围：[2,6)随机整数
int_arr4 = np.random.randint(2,6,size=(2,3))
print('low=2,high=6,size=(2,3)：生成一个2*3整数数组:',int_arr4)
```

```python
#  numpy.random.randn(d0, d1, ..., dn)：生成一个浮点数或N维浮点数组 —— 正态分布
import matplotlib.pyplot  as plt
%matplotlib inline

samples1 = np.random.randn(1000)
samples2 = np.random.randn(1000)
plt.scatter(samples1,samples2)
```
